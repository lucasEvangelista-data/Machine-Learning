**Aplicando o Modelo de Regressão Linear**  
Vamos agora começar a treinar o modelo de regressão! Primeiro, precisamos dividir nossos dados em uma matriz X que contém os recursos para treinamento e uma matriz y com a variável de destino, neste caso a coluna Preço. Jogaremos fora a coluna Endereço, pois ela possui apenas informações de texto que o modelo de regressão linear não pode usar.  
Matriz X e y  
**X - variáveis preditoras**  
**Y - variável de resposta**

```python
X = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms','Avg. Area Number of Bedrooms', 'Area Population']]
y = df['Price']
X.head()
```

```python
y.head()
```

**Dividindo os dados em treino e teste**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=50)
```

```python
from sklearn.linear_model import LinearRegression

# Aqui criamos o modelo preditor
modelopreditor = LinearRegression()

# Aqui ocorre o aprendizado
modelopreditor.fit(X_train,y_train)
```

```python
# Imprimindo os coeficientes com os nomes das colunas
coeff = pd.DataFrame(modelopreditor.coef_,X.columns,columns=['Coeficiente'])
coeff
```

```python
# fazendo predições com os dados de teste
predictions = modelopreditor.predict(X_test)

# imprimindo as primeiras variáveis preditoras
X_test.head()
```

```python
# Predições realizadas
predictions
```

```python
# Valores reais para acertar
y_test.head()
```

**AVALIANDO O MODELO**  
Aqui estão três métricas de avaliação comuns para problemas de regressão:  
• **Mean Absolute Error (MAE)** é a média do valor absoluto dos erros:  
• **Mean Squared Error (MSE)** é a média dos erros quadráticos:  
• **Root Mean Squared Error (RMSE)** é a raiz quadrada da média dos erros ao quadrado:  
Comparando estas métricas:  
• MAE é o mais fácil de entender, porque é o erro médio.  
• MSE é mais popular que o MAE, porque o MSE "pune" erros maiores, o que tende a ser útil no mundo real.  
• RMSE é ainda mais popular que o MSE, porque o RMSE é interpretável nas unidades do eixo "y".  
Todas essas são funções de perda, porque queremos minimizá-las.  
MAPE: erro médio absoluto em porcentagem

Para uma explicação sobre as várias métricas de avaliação, veja o artigo Métricas para Regressão: Entendendo as métricas R², MAE, MAPE, MSE e RMSE:

https://medium.com/data-hackers/prevendo-n%C3%BAmeros-entendendo-m%C3%A9tricas-de-regress%C3%A3o-35545e011e70


```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
print('MAPE:', metrics.mean_absolute_percentage_error(y_test, predictions))
print('R2:', metrics.r2_score(y_test, predictions))
```
